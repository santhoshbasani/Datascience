```python
flight_data = pd.read_csv(r"C:\Users\kalya\Santhosh\Data Science\Notebook\Statistics\flight_delays.csv", index_col = 'Month')
```

```python
flight_data.head()
```

### bar graph

```python
plt.figure(figsize=(16,8))
plt.title('Average arrival delays for the airline WN, by month') 

#BAR GRAPH
sns.barplot(x=flight_data.index, y = flight_data['WN']) 

#Add the labels
plt.xlabel('Months')
plt.ylabel('Arrival delays')
```

### Heatmap

```python
plt.figure(figsize=(10,8))
plt.title('Average arrival delays for the airline WN, by month') 

#BAR GRAPH
sns.heatmap(data = flight_data) 

#Add the labels
plt.xlabel('Months')
plt.ylabel('Arrival delays')
```

```python
plt.figure(figsize=(10,8))
plt.title('Average arrival delays for the airline WN, by month') 

#BAR GRAPH
sns.heatmap(data = flight_data, annot=True) 

#Add the labels
plt.xlabel('Months')
plt.ylabel('Arrival delays')
```

### above long white bar is having null values for US

```python
plt.figure(figsize=(10,8))
plt.title('Average arrival delays for the airline WN, by month') 

#BAR GRAPH
sns.heatmap(data = flight_data, annot=True, cmap = 'YlGnBu') 

#Add the labels
plt.xlabel('Months')
plt.ylabel('Arrival delays')
```

```python
plt.figure(figsize=(10,8))
plt.title('Average arrival delays for the airline WN, by month') 

#BAR GRAPH
sns.heatmap(data = flight_data, annot=True, cmap = 'GnBu') 

#Add the labels
plt.xlabel('Months')
plt.ylabel('Arrival delays')
```

```python
flight_data.corr()
```

```python
plt.figure(figsize=(10,8))
plt.title('Average arrival delays for the airline WN, by month') 

#Heat map
corr = flight_data.corr()
sns.heatmap(data = corr, annot=True, cmap = 'PRGn') 

#Add the labels
plt.xlabel('Months')
plt.ylabel('Arrival delays')
```
