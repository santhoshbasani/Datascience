### get_dummies
### convert the dataframe for the specified columns and add those convertable columns into the existing dataframe and remove the original column after converting the columns
### The main reason to make the columns to convertable is for the mathematical calculation in the model envrionments.  This is all part of the preprocessing on the data.  
###
### In the below example, First get all the object type column names into list.
### Get each object type column name from list and make it convertable column using get_dummies by prefixing the existing
### column name for the new column name
### once we make the convertable column on the existing column, then drop the existing column as we already done the convertabel column
### Now add the convertable column to the existing dataframe with the option axis = 1 to add the columns

```python
def dummy_df(df, dummy_col_list):
    for x in dummy_col_list:
        dummies = pd.get_dummies(df[x], prefix = x, dummy_na = False)
        df = df.drop(x, 1)
        df = pd.concat([df, dummies], axis = 1)
    return df
```

```python
x = dummy_df(x, object_columns)
```

### Haingling missing values

### SimpleImputer helps to replace the null values with median for every single column
It involves two steps

First is to fit the imputer to the existing dataframe (learning process)

Second, we tranform the learned/fitted imputer to a dataframe

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='median')
imp.fit(x)
```

```python
x = pd.DataFrame(imp.transform(x), columns=x.columns)
```
